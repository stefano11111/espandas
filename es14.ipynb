```python
# Write a Pandas program to display the dataframe in Heatmap style.

import pandas as pd
import random
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns



db=pd.DataFrame({random.randint(0,100):[random.randint(0,101) for _ in range (10)],
                 random.randint(0,100):[random.randint(0,101) for _ in range (10)],
                 random.randint(0,100):[random.randint(0,101) for _ in range (10)],
                 random.randint(0,100):[random.randint(0,101) for _ in range (10)],
                 })
l=random.randint(5,15)
for x in range(l):
    a=random.randint(1,9)
    b=random.randint(0,3)
    db.iloc[a,b]=np.nan
print(db)

```

```python
sns.heatmap(db,annot=True,center=0)
    
```
