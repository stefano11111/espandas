```python
# Write a Pandas program to display the dataframe in Heatmap style.

import pandas as pd
import random
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns



db=pd.DataFrame({"col1":[random.randint(0,11) for _ in range (10)],
                 "col2":[random.randint(0,11) for _ in range (10)],
                 "col3":[random.randint(0,11) for _ in range (10)],
                 "col4":[random.randint(0,11) for _ in range (10)],
                 })
l=random.randint(5,15)
for x in range(l):
    a=random.randint(1,9)
    b=random.randint(0,3)
    db.iloc[a,b]=np.nan
print(db)
```

```python
def highlight_nan_column(s):
    
    is_nan_column = s.isna().any()
    
    return ['background-color: red' if is_nan_column else '' for _ in s]

styled_df = db.style.apply(highlight_nan_column)

styled_df
```
